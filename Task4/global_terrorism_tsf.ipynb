# Research insights using plot

<h3>Plotting based on region name</h3>

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.subplots(figsize=(25, 5))
sns.countplot(x='region_txt', data=terrorism, palette='inferno', order=terrorism['region_txt'].value_counts().index)
```

# Plotting based on year of attack

```python
plt.subplots(figsize=(30, 8))
sns.countplot(x='iyear', data=terrorism, palette='RdYlGn_r', edgecolor=sns.color_palette('dark', 7))
plt.xlabel('Year of attack')
plt.title('No. of terrorist activities each year')
```

# Plotting based on attacktype

```python
plt.subplots(figsize=(25, 8))
sns.countplot(x='attacktype1_txt', data=terrorism, order=terrorism['attacktype1_txt'].value_counts().index)
plt.xlabel('Attack type')
plt.title('Attacking Methods by Terrorists')
```

# Relationship between different factors

```python
terrorism_region = pd.crosstab(terrorism.iyear, terrorism.region_txt)
terrorism_region.plot(color=sns.color_palette('Set2', 12))
fig = plt.gcf()
fig.set_size_inches(13, 6)
```

```python
terrorism_targ_type = pd.crosstab(terrorism.iyear, terrorism.targtype1_txt)
terrorism_targ_type.plot(color=sns.color_palette('Set2', 12))
fig= plt.gcf()
fig.set_size_inches(18,10)
```

```python
plt.subplots(figsize=(28,8))
terrorism.groupby(['region_txt', 'iyear']).count()['casualties'].plot()
plt.ylabel('Casualties')
```
